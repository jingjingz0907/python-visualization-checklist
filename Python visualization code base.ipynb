**5. Heatmap**

```python
# Generate a 30x30 correlation matrix 
df_corr = df.corr()

# Generate a mask for the upper triangle
mask = np.zeros_like(df_corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(df_corr, mask=mask, cmap=cmap, vmax=.3, center=0, 
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

```python
file = open("constitution.txt")
constitution = file.read()
constitution_words = constitution.split()
constitution_words[0:10]
```

**7. Wordcloud**

```python
# Python program to generate WordCloud
 
# importing all necessery modules
from wordcloud import WordCloud, STOPWORDS
 
comment_words = ' '
stopwords = set(STOPWORDS)
 
# iterate through the csv file
for val in constitution_words:
     
    # typecaste each val to string
    val = str(val)
 
    # split the value
    tokens = val.split()
     
    # Converts each token into lowercase
    for i in range(len(tokens)):
        tokens[i] = tokens[i].lower()
         
    for words in tokens:
        comment_words = comment_words + words + ' '
 
 
wordcloud = WordCloud(width = 800, height = 800,
                background_color ='white',
                stopwords = stopwords,
                min_font_size = 10).generate(comment_words)
 
# plot the WordCloud image                       
plt.figure(figsize = (8, 8), facecolor = None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad = 0)
```

**8. Barplot**

```python
music = pd.read_csv('Music Sales.csv')
music.head()
```

```python
list_of_values = ['Europe', 'Asia']
asia_euro = music[music['Region'].isin(list_of_values)]
asia_euro = asia_euro.groupby('Genre').sum().sort_values(by = 'Quantity', ascending = False)
asia_euro.head()
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
ax = sns.barplot(x="Quantity", y=asia_euro.index, data=asia_euro, hue = asia_euro.index, dodge = False)
```

**9. Stacked barplot**

```python
artist = music.groupby('Artist').sum().sort_values(by = 'Quantity', ascending = False).iloc[:20].index
music_20artist = music[music['Artist'].isin(artist)]
table = pd.pivot_table(music_20artist, values = 'Quantity', index = ['Artist'], columns = ['Genre'], aggfunc=np.sum)
table.head()
```

```python
ax = table.plot.barh(stacked=True, figsize=(25, 15), cmap=plt.cm.get_cmap('tab10'))
plt.title('Top 20 Artists in Asia and Europe', size = 25)
plt.xlabel('Quantity')
```

**10. Line plot**

```python
music["InvoiceDate"] = pd.to_datetime(music['InvoiceDate'])
music["InvoiceDate"] = pd.to_datetime(music.InvoiceDate.dt.date).dt.to_period('M').dt.to_timestamp()
Genrelist = ['Alternative & Punk', 'Heavy Metal', 'Rock']
music = music[music['Genre'].isin(Genrelist)]
music.head()
```

```python
Salestable = pd.pivot_table(music, values = 'Sales', index = ['InvoiceDate'], columns = ['Genre'], aggfunc=np.sum)
Salestable = Salestable.reset_index()
```

```python
plt.rcParams['figure.figsize'] = (25, 10)
ax = sns.lineplot(x=Salestable['InvoiceDate'], y=Salestable['Rock'], label='Rock', linewidth = 5)
ax = sns.lineplot(x=Salestable['InvoiceDate'], y=Salestable['Heavy Metal'], label='Heavy Metal', linewidth = 5)
ax = sns.lineplot(x=Salestable['InvoiceDate'], y=Salestable['Alternative & Punk'], label = 'Alternative & Punk', linewidth = 5)

ax.set(xlabel='date', ylabel='sales')
ax.xaxis.label.set_size(20)
ax.yaxis.label.set_size(20)
plt.setp(ax.get_legend().get_texts(), fontsize='22')
```

**11. Pie chart**

```python
# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [15, 30, 45, 10]
explode = (0, 0.1, 0, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```
